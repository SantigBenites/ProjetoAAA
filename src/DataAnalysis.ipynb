```python
import matplotlib.pyplot as plt
import json
  
# reading the data from the file
with open('../assets/data.txt') as f:
    data = f.read()
data = json.loads(data)


episode = data["episode"]
duration = data["duration"]
step_num = data["step_num"]
GamesWonByBlack = data["GamesWonByBlack"]
GamesWonByWhite = data["GameWonByWhite"]

def add_if_dont_exist(data_set:dict,key:str,val:str):

    if key in data_set.keys():
        data_set[key].append(val)
    else:
        data_set[key] = [val]

def Average(lst):
    return sum(lst) / len(lst)

duration_data = {}
step_num_data = {}
for index,ep in enumerate(episode):
    add_if_dont_exist(duration_data,int(ep),int(duration[index]))
    add_if_dont_exist(step_num_data,int(ep),int(step_num[index]))


for  key in duration_data.keys():
    av = Average(duration_data[key])
    duration_data[key] = av

for  key in step_num_data.keys():
    av = Average(step_num_data[key])
    step_num_data[key] = av

plt.plot(duration_data.keys(), duration_data.values())
plt.xlabel("Episode")
plt.ylabel("Duration")
plt.show()


plt.plot(step_num_data.keys(), step_num_data.values())
plt.xlabel("Episode")
plt.ylabel("Steps")
plt.show()


courses = ["White","Black"]
values = [1123, 877]

plt.bar(courses, values, color ='maroon',width = 0.4)

plt.title("Games won by each player")
plt.show()


courses = ["RL","SF"]
values = [56, 44]

plt.bar(courses, values, color ='maroon',width = 0.4)

plt.title("Games won by each player")
plt.show()


courses = ["RL","GA"]
values = [37, 63]

plt.bar(courses, values, color ='maroon',width = 0.4)

plt.title("Games won by each player")
plt.show()
```
